```python
from ecg_qc import EcgQc
import pandas
from pandas import *
import os
from openpyxl import load_workbook
import numpy as np
from ecgdetectors import Detectors
import biosppy.signals.ecg as bsp_ecg

ecg_qc = EcgQc()
ecg_qc = EcgQc('rfc_norm_2s.pkl',
               sampling_frequency=256,
               normalized=False)

sampling_frequency= 500
detectors = Detectors(sampling_frequency)

def ecg_quality_check(path):
    for root, _, file_list in os.walk(path):
        print()
        print()
        print("In directory {}".format(root))
        print()
        for filename in file_list:
            os.chdir(root)
            # reading CSV file
            data = read_csv(filename)
            # converting column data to list
            column1 = data['1'].tolist()
            column6 = data['6'].tolist()

            signal_quality = ecg_qc.get_signal_quality(column1)
            sqi_scores = ecg_qc.compute_sqi_scores(column1)
            signal_quality = ecg_qc.predict_quality(sqi_scores)
            qrs_frames_hamilton = bsp_ecg.hamilton_segmenter(signal=np.array(column1),sampling_rate=sampling_frequency)[0]
            print("File: ",filename)
            print("Lead: 1","|","Signal Quality: ",signal_quality,"|","SQI Scores: ",sqi_scores)
            if(signal_quality == 1):
                print("Hamilton Frames: ",qrs_frames_hamilton)
            print()

            signal_quality = ecg_qc.get_signal_quality(column6)
            sqi_scores = ecg_qc.compute_sqi_scores(column6)
            signal_quality = ecg_qc.predict_quality(sqi_scores)
            qrs_frames_hamilton = bsp_ecg.hamilton_segmenter(signal=np.array(column6),sampling_rate=sampling_frequency)[0]
            print("Lead: 6","|","Signal Quality: ",signal_quality,"|","SQI Scores: ",sqi_scores)
            if(signal_quality == 1):
                print("Hamilton Frames: ",qrs_frames_hamilton)

            print("____________________________________________________________________________________________________________")
```
